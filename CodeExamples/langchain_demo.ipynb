```python
import os

from langchain.chat_models import ChatOpenAI
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain

def get_response(llm, template, input_data):
    
    prompt = PromptTemplate(template=template, input_variables=["question"])
    llm_chain = LLMChain(prompt=prompt, llm=llm)
    response = llm_chain.run(input_data)

    return response

```

```python
llm = ChatOpenAI(temperature=0.1, model='gpt-3.5-turbo')
template = """Sentiment analysis: {question}"""
question = "I just had the best day of my life at the amusement park!"

input_data = {
    "question": question   
}

response = get_response(llm, template, input_data)
print(response)
```
