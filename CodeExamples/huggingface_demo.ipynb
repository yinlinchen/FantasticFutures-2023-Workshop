```python
import os
from dotenv import load_dotenv

load_dotenv()
```

```python
from langchain.llms import HuggingFaceHub
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain


def get_response(llm, template, input_data):
    
    prompt = PromptTemplate(template=template, input_variables=["question"])
    llm_chain = LLMChain(prompt=prompt, llm=llm)
    response = llm_chain.run(input_data)

    return response

```

```python
repo_id = "google/flan-t5-xxl"

llm = HuggingFaceHub(
    repo_id=repo_id, model_kwargs={"temperature": 0.1, "max_length": 200}
)

template = """Question: {question}

Answer: Let's think step by step."""

question = "Where is the capital of Canada?"

input_data = {
    "question": question   
}

response = get_response(llm, template, input_data)
print(response)
```

```python
template = "Sentiment analysis for {question}"
question = "I absolutely love the new design. It's innovative, user-friendly, and visually stunning!"

input_data = {
    "question": question   
}

response = get_response(llm, template, input_data)
print(response)
```
